```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib
import requests
import re
from tabulate import tabulate
```

```python
DOMAIN_ROOT = 'http://www.basketball-reference.com'
response = requests.get('https://www.basketball-reference.com/leagues/NBA_2018_games.html')
page = response.text
```

```python
soup = BeautifulSoup(page)
```

```python
df = pd.DataFrame()
for month in soup.select('#content > div[class=filter] > div > a'):
    month_txt = str(DOMAIN_ROOT + month['href'])
    month_response = requests.get(month_txt)
    month_page = month_response.text
    month_soup = BeautifulSoup(month_page)
    for box in month_soup.select('#schedule > tbody > tr > td[data-stat=box_score_text] > a'):
        box_txt = str(DOMAIN_ROOT + box['href'])
        box_response = requests.get(box_txt)
        box_page = box_response.text
        box_soup = BeautifulSoup(box_page.content)
        table = soup.find_all('table')
        df = pd.read_html(str(table))
        data1 = pd.DataFrame(df[0][:-1])
        data2 = pd.DataFrame(df[8][:-1])
        data1.columns = data1.columns.droplevel()
        data2.columns = data2.columns.droplevel()
        data1 = data1.drop([5]).reset_index()
        data1 = data1.drop('index', axis=1)
        data2 = data2.drop([5]).reset_index()
        data2 = data2.drop('index', axis=1)
        df = pd.concat([df, data1, data2])





#         for key in box_soup.find_all(class_="box > h1"):
#         for tables in box_soup.find_all(class_="sortable"):
#             for row in tables.select_one("tbody").find_all("tr", class_=lambda x: x != 'thread'):
#                 for starter in row.select('th'):
#                     array = []
#                     name = starter.text
#                     for td in row.select("td"):
#                         array.append(td.text)
#                         temp_dict.update({name: array})
#           dictionary.update({})
#       print(dictionary)
# dictionary
```
